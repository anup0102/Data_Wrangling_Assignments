Filter the dataframe with only this columns 'continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index' and update the data frame.

```python
covid = covid[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

Remove duplicated

```python
covid.drop_duplicates()
```

mising values from all columns

```python
covid.isnull()
```

remove observations where continent value is missing

```python
covid.dropna(subset=['continent'])
```

fill all missing values with 0

```python
covid.fillna(0)
```

Convert date column in datetime format using pandas.to_datetime

```python
covid['date'] = covid['date'].apply(pd.to_datetime)
```

Create new column month after extracting month data from date column.

```python
import datetime
covid['month'] = pd.DatetimeIndex(covid['date']).month
```

Find max value in all columns using groupby function on 'continent' column

```python
 df_groupby = covid.groupby('continent').max().reset_index()
```

```python
df_groupby
```

8. Feature Engineering :

 a. Create a new feature 'total_deaths_to_total_cases' by ratio of  'total_deaths' column to 'total_cases'

```python
covid['total_deaths_to_total_cases'] = covid['total_deaths']/covid['total_cases']
```
